```python
import pandas as pd
import json
```

```python
# errors from active presentations
with open(r'C:\Users\Marcelo\Desktop\Medicamentos\extracao-dados-medicamentos\data_pipelines\dags\temp_files\2025-05-10presentations_from_active_medicines.json', 'r', encoding="utf8") as f:
    ativos = json.load(f)
```

```python
with open(r'C:\Users\Marcelo\Desktop\Medicamentos\extracao-dados-medicamentos\data_pipelines\dags\temp_files\2025-05-10presentations_from_inactive_medicines.json', 'r', encoding="utf8") as f:
    inativos = json.load(f)
```

```python
apresentacoes_ativos = []
for a in ativos:
    aps = a['apresentacoes']
    for ap in aps:
        ap['codigo_medicamento'] = a['codigoProduto']
        ap['codigo_notificacao_medicamento'] = a['codigoNotificacao']
        ap['nome_comercial_medicamento'] = a['nomeComercial']
    apresentacoes_ativos += aps
del ativos
```

```python
apresentacoes_inativos = []
for a in inativos:
    aps = a['apresentacoes']
    for ap in aps:
        ap['codigo_medicamento'] = a['codigoProduto']
        ap['codigo_notificacao_medicamento'] = a['codigoNotificacao']
        ap['nome_comercial_medicamento'] = a['nomeComercial']
    apresentacoes_inativos += aps
del inativos
```

```python
acondicionamentos_ativos = []
for a in ativos:
    aps = a['acondicionamentos']
    for ap in aps:
        ap['codigo_medicamento'] = a['codigoProduto']
        ap['codigo_notificacao_medicamento'] = a['codigoNotificacao']
        ap['nome_comercial_medicamento'] = a['nomeComercial']
    acondicionamentos_ativos += aps
del ativos
```

```python
acondicionamentos_inativos = []
for a in inativos:
    aps = a['acondicionamentos']
    for ap in aps:
        ap['codigo_medicamento'] = a['codigoProduto']
        ap['codigo_notificacao_medicamento'] = a['codigoNotificacao']
        ap['nome_comercial_medicamento'] = a['nomeComercial']
    acondicionamentos_inativos += aps
del inativos
```

```python
len(apresentacoes_inativos)
```

```python
apresentacoes_ativos = pd.DataFrame(apresentacoes_ativos)
apresentacoes_ativos['medicamento_ativo'] = True
```

```python
acondicionamentos_ativos = pd.DataFrame(acondicionamentos_ativos)
acondicionamentos_ativos['medicamento_ativo'] = True
```

```python
apresentacoes_inativos = pd.DataFrame(apresentacoes_inativos)
apresentacoes_inativos['medicamento_ativo'] = True
```

```python
acondicionamentos_inativos = pd.DataFrame(acondicionamentos_inativos)
acondicionamentos_inativos['medicamento_ativo'] = True
```

```python
apresentacoes = pd.concat([apresentacoes_ativos, apresentacoes_inativos], ignore_index=True)
del apresentacoes_ativos
del apresentacoes_inativos
```

```python
acondicionamentos = pd.concat([acondicionamentos_ativos, acondicionamentos_inativos])
del acondicionamentos_ativos
del acondicionamentos_inativos
```

```python
apresentacoes.to_excel("Apresentações Medicamentos.xlsx")
```
