```python
medicines_df = df.copy()

medicines_df = pd.concat([
    medicines_df.drop(columns=['produto', 'empresa', 'processo']),
    pd.json_normalize(medicines_df['produto']),
    pd.json_normalize(medicines_df['empresa']),
    pd.json_normalize(medicines_df['processo']),
], axis=1)
medicines_df = medicines_df[medicines_df["registro_ativo"]].drop(columns=["registro_ativo"])

data_perf = []
for column in medicines_df.columns:
    examples = medicines_df[medicines_df[column].notnull()][column]
    perf = {
        "column": column,
        "type": type(examples.iloc[-1:].item()) if len(examples) else None,
        "example": examples.iloc[-1:].item() if len(examples) else None,
        "filled_values": len(medicines_df[medicines_df[column].notnull()]),
        "null_values": len(medicines_df[medicines_df[column].isnull()]),
        "unique_values": len(medicines_df[medicines_df[column].notnull()][column].unique()),
        "duplicated_values": len(medicines_df[medicines_df[column].notnull() & medicines_df[column].duplicated()]),
    }
    data_perf.append(perf)
data_perf = pd.DataFrame(data_perf)
data_perf["type"] = data_perf["type"].apply(lambda x: "STRING" if x == str else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "INTEGER" if x == int else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "FLOAT" if x == float else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "BOOLEAN" if x == bool else x)
```

```python
pd.DataFrame(data_perf).to_csv("Data Perf Registros Ativos.csv", encoding="utf-8")
```

```python
analyses_df = df.copy()

analyses_df = pd.concat([
    analyses_df.drop(columns=['empresa', 'processo', 'processoMatriz']),
    pd.json_normalize(analyses_df['empresa'], sep='.').add_prefix('empresa.'),
    pd.json_normalize(analyses_df['processo'], sep='.').add_prefix('processo.'),
    pd.json_normalize(analyses_df['processoMatriz'], sep='.').add_prefix('processoMatriz.'),
], axis=1)

data_perf = []
for column in analyses_df.columns:

    _column= column
    examples = analyses_df[analyses_df[column].notnull()][column]
    _type= type(examples.iloc[-1:].item()) if len(examples) else None

    analyses_df[column] = analyses_df[column].apply(lambda x: None if x==[] else x)
    examples = analyses_df[analyses_df[column].notnull()][column]
    
    _example = examples.iloc[-1:].item() if len(examples) else None
    _filled_values = len(analyses_df[analyses_df[column].notnull()])
    _null_values = len(analyses_df[analyses_df[column].isnull()])
    _unique_values = len(analyses_df[analyses_df[column].notnull()][column].unique()) if _type not in (list, dict) else "-"
    _duplicated_values = len(analyses_df[analyses_df[column].notnull() & analyses_df[column].duplicated()]) if _type not in (list, dict) else "-"

    perf = {
        "column": _column,
        "type": _type,
        "example": _example,
        "filled_values": _filled_values,
        "null_values": _null_values,
        "unique_values": _unique_values,
        "duplicated_values": _duplicated_values
    }
    data_perf.append(perf)
data_perf = pd.DataFrame(data_perf)
data_perf["type"] = data_perf["type"].apply(lambda x: "STRING" if x == str else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "INTEGER" if x == int else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "FLOAT" if x == float else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "BOOLEAN" if x == bool else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "LIST" if x == list else x)
data_perf["type"] = data_perf["type"].apply(lambda x: "MAP" if x == dict else x)
```

```python
pd.DataFrame(data_perf).to_csv("Preço Máximo ao Governo.csv", encoding="utf-8")
```
