```python
from seleniumbase import Driver
import time

# Create a new instance of the Chrome driver (or Firefox)
driver = Driver(uc=True, uc_cdp_events=True)

# Function to add headers to all requests
def add_headers(headers):
    driver.execute_cdp_cmd('Network.enable', {})
    driver.execute_cdp_cmd('Network.setExtraHTTPHeaders', {'headers': headers})

# Enable request/response logging
def log_network_activity():
    driver.execute_cdp_cmd("Network.enable", {})
    driver.execute_cdp_cmd("Network.setRequestInterception", {"patterns": [{"urlPattern": "*"}]})

# Listen to network events and print headers
def print_headers(request):
    headers = request.get("request", {}).get("headers", {})
    print("Request Headers:", headers)

# Add the "Authorization: Guest" header to all requests
add_headers({"Authorization": "Guest"})

# Open the first URL
driver.get('https://consultas.anvisa.gov.br')
time.sleep(1.2)

# Attach listener for network events
log_network_activity()
driver.add_cdp_listener("Network.requestWillBeSent", print_headers)

# Open the second URL with the custom header
driver.get('https://consultas.anvisa.gov.br/api/consulta/medicamento/produtos/?column=&count=10&filter%5BsituacaoRegistro%5D=V&order=asc&page=1')
time.sleep(4)

# Close the browser
driver.quit()
```
